```python
# extra data needed to run the engine 
data_points = ['Total common shares outstanding', 'Float shares outstanding',
#--------------- [0 - 1] -------------
 'Number of employees', 'Number of shareholders', 
#---------------- [2,3] ---------------------
'Price to earnings ratio', 'Price to sales ratio', 
# ------------------ [4,5] -------------------------------
'Price to cash flow ratio', 'Price to book ratio', 'Enterprise value',
#------------------[6,7,8] -----------------------------------
 'Enterprise value to EBITDA ratio', 'Return on assets %', 'Return on equity %',
# -----------------------[9,10] --------------------------------
 'Return on invested capital %', 'Gross margin %', 
# -----------------------[11,12,13,14,15] -----------------------
'Operating margin %', 'EBITDA margin %', 'Net margin %', 'Quick ratio', 'Current ratio',
# -----------------------[16,17,18] ---------------------------
 'Inventory turnover', 'Asset turnover', 'Debt to assets ratio',
#------------------------ [19,20] --------------------------------
 'Debt to equity ratio', 'Long term debt to total assets ratio']

def plot_data(point,company_data,company_name):  

    fig =plt.figure()
    fig,axes = plt.subplots(1,1,figsize=(15,5))
    axes.grid(True)
    axes.plot(company_data[data_points[point]],label=company_name)
    axes.set_ylabel(data_points[point])
    axes.set_title(company_name)
    axes.set_xlabel("year")
    axes.legend(loc = "best")

  
    return axes

def compare_companies(point,company_data_1,company_name_1,company_data_2,company_name_2):

    fig =plt.figure()
    fig,axes = plt.subplots(1,1,figsize=(15,5))
    axes.grid(True)
    axes.plot(company_data_1[data_points[point]],label=company_name_1)
    axes.plot(company_data_2[data_points[point]],label=company_name_2)

    axes.set_ylabel(data_points[point])
    
    axes.set_xlabel("year")
    axes.legend(loc = "best")
```
