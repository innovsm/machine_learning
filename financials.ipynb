```python
import numpy as np
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from resources import income_stat,trimmer,url_maker
import matplotlib.pyplot as plt
```

```python

def raw_data(url_list):
    main_list = []

    for i in url_list:
        driver = webdriver.Chrome("C:\webdriver\chromedriver.exe")
        driver.get(i)
        data_1 = []
        titles = driver.find_elements_by_class_name('value-Coz8DukZ')
        for t in titles:
            data_1.append(t.text)
            
       
        print(data_1)  
        main_list.append(data_1)
    return main_list

def value_maker(value):
    x1= value.split("\u202a")
    y1 = x1[1].split("\u202c")
    if(value.count("B") > 0):
        z1 = y1[0].split("B")
        if("−" in z1[0]):
            data = float(z1[0].replace("−","-"))*1000000000
        else:
            data = float(z1[0])*1000000000
    elif(value.count("M") > 0):
        z1 = y1[0].split("M")
        if("−" in z1[0]):
            data = float(z1[0].replace("−","-"))*1000000
        else:
            data = float(z1[0])*1000000
    elif(value.count("T")):
        z1 = y1[0].split("T")
        if("−" in z1[0]):
            data = float(z1[0].replace("−","-"))*1000000000000
        else:
            data = float(z1[0])*1000000000000
    else:
        data = float(y1[0])
    
    return data
```

```python
def income_statment(company_name):
    company_list = [company_name]
    x1 = raw_data([url_maker(i,1) for i in company_list])   # change company here
    count = 0
    main_count = 0
    super_list = []
    main_list = []
    for i in range(len(x1[0])):
        main_list.append(x1[0][main_count])
        main_count += 1
        count += 1
        if(count == 8):
            count = 0
            super_list.append(main_list)
            main_list = []
    data_1 = pd.DataFrame(super_list)
    data_1.columns = data_1.iloc[0]
    data_1.drop([0], inplace = True)
    data_1
    X= data_1.drop([11,13,15,16,17,18,19,20,8,10,12])
    X.index = ["Gross Profit", "Total Revenue","COGS", "Operating Expenses", "Operating Income", "Non-Operating Income", "Pretax Income",  "Taxes",  "Net Income",  "EBITDA", "EBIT", "Total Operating Expenses"]
    for i in range(8):
        X.loc["COGS"][i] = value_maker(X.loc["COGS"][i])
        X.loc["Operating Expenses"][i] = value_maker(X.loc["Operating Expenses"][i])
        X.loc["Operating Income"][i] = value_maker(X.loc["Operating Income"][i])
        X.loc["Gross Profit"][i] = value_maker(X.loc["Gross Profit"][i])
        X.loc["Taxes"][i] = value_maker(X.loc["Taxes"][i])
        X.loc["EBIT"][i] = value_maker(X.loc["EBIT"][i])
        X.loc["EBITDA"][i] = value_maker(X.loc["EBITDA"][i])
        X.loc["Total Operating Expenses"][i] = value_maker(X.loc["Total Operating Expenses"][i])
        X.loc["Total Revenue"][i] = value_maker(X.loc["Total Revenue"][i])
        X.loc["Net Income"][i] = value_maker(X.loc["Net Income"][i])
        X.loc["Non-Operating Income"][i] = value_maker(X.loc["Non-Operating Income"][i])
        X.loc["Pretax Income"][i] = value_maker(X.loc["Pretax Income"][i])
    return X
```
